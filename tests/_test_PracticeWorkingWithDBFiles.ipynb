```python
db_dir = "/Users/jackbarlow/Dropbox/Jack B/data/ISBT/Stencil mask 10L & EBL 11L/bluefors vector 20221111/databases"
db_files_list = []
if os.path.isdir(db_dir):
    for file in os.listdir(db_dir):
        if file.endswith('.db'):
            db_files_list.append(file)
db_files_list.sort()
db_files = pandas.DataFrame({'file': db_files_list})
db_files
```
